```python
## get the trainandvalid data set 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

bull_dozer = pd.read_csv("data/TrainAndValid.csv",
                         low_memory=False,
                         parse_dates=["saledate"])
bull_dozer.head().T
```

```python
## reduce the data 

bull_dozer = bull_dozer.iloc[:200000]
len(bull_dozer)
```

```python
## check for missing values and check the datatypes

bull_dozer.isna().sum()
```

```python
bull_dozer.dtypes
```

```python
## sort column saledateby date 

bull_dozer.sort_values(by=["saledate"],inplace=True,ascending=True)
```

```python
bull_dozer.saledate.head()
```

```python
## add own feature ie add own columns 

bull_dozer["saleyear"] = bull_dozer["saledate"].dt.year
bull_dozer["salemonth"] = bull_dozer["saledate"].dt.month
bull_dozer["saleday"] = bull_dozer["saledate"].dt.day
bull_dozer["saledayofweek"] = bull_dozer["saledate"].dt.dayofweek
bull_dozer["saledayofyear"] = bull_dozer["saledate"].dt.dayofyear

bull_dozer.drop("saledate",axis=1,inplace=True)
```

```python
bull_dozer.saledate
```

```python
bull_dozer.head().T
```

```python
# check whether the columns are of string datatype if string convert them to category 


for label,content in bull_dozer.items():
    if(pd.api.types.is_string_dtype(content)):
        print(label)
        ## convert to category 
        bull_dozer[label] = content.astype('category').cat.as_ordered()
```

```python
bull_dozer.dtypes
```

```python
## check the numeric columns and fill thier missing values 

for label,content in bull_dozer.items():
    if(pd.api.types.is_numeric_dtype(content)):
        print(label)
```

```python
for label,content in bull_dozer.items():
    if(pd.api.types.is_numeric_dtype(content)):
        if(pd.isnull(content).sum()):
            print(label)
```

```python
# fill the missing values with median 

for label,content in bull_dozer.items():
    if(pd.api.types.is_numeric_dtype(content)):
        # add a binary column shows whether it had missing values or not 
        bull_dozer[label + "_missing"] = pd.isnull(content)
        # fill the missing values 
        bull_dozer[label] = content.fillna(content.median())
        
    # fill categorical column and turn into numbers
    if not(pd.api.types.is_numeric_dtype(content)):
        # add a binary columnn 
        bull_dozer[label+"_missing"] = pd.isnull(content)
        # fill and turn values 
        bull_dozer[label] = pd.Categorical(content).codes + 1
```

```python
bull_dozer.isna().sum()
```

```python
bull_dozer.head().T
```

## MOdelling 

```python
%%time
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(max_samples=1000)

# fit the model 

model.fit(bull_dozer.drop("SalePrice",axis=1),bull_dozer["SalePrice"])
```

```python
model.score(bull_dozer.drop("SalePrice",axis=1),bull_dozer["SalePrice"])
```

```python
## get train and valid split 

bulldozer_valid = bull_dozer[bull_dozer.saleyear==2009]
bulldozer_train = bull_dozer[bull_dozer.saleyear!=2009]
```

```python
bulldozer_valid.shape, bulldozer_train.shape
```

```python
# split into Xtrain and Xvalid 

X_valid,y_valid = bulldozer_valid.drop("SalePrice",axis=1),bulldozer_valid["SalePrice"]
X_train,y_train = bulldozer_train.drop("SalePrice",axis=1),bulldozer_train["SalePrice"]
```

```python
X_valid.shape,y_valid.shape,X_train.shape,y_train.shape
```

```python
# fit the model 

model.fit(X_train,y_train)
```

```python
# create a method to evaluate a score we need root-mean-squared-log-error 

from sklearn.metrics import mean_squared_log_error,mean_absolute_error,r2_score

def rmsle(y_true,y_preds):
    """
    calculate root mean squareed log error between true labels and predicted labels 
    """
    return np.sqrt(mean_squared_log_error(y_true,y_preds))

def show_scores(model):
    # make predictions 
    train_preds = model.predict(X_train)
    valid_preds = model.predict(X_valid)
    
    scores = {"trainingMSE":mean_absolute_error(y_train,train_preds),
              "validMSE":mean_absolute_error(y_valid,valid_preds),
              "trainingRMSLE":rmsle(y_train,train_preds),
              "ValidRMSLE":rmsle(y_valid,valid_preds),
              "Train_r2_":r2_score(y_train,train_preds),
              "Valid_r2":r2_score(y_valid,valid_preds)}
    return scores
```

```python
# evaluate score 

show_score(model)
```

```python
# fit the model 

model.fit(X_train,y_train)
```

```python
show_scores(model)
```

## Tuning hyperparameters using RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

rs_grid = {"n_estimators":np.arange(10,100,10),
           "max_features":[0.5,1.0,"sqrt","auto"],
           "min_samples_split":np.arange(1,20,2),
           "min_samples_leaf":np.arange(1,20,2),
           "max_samples":[1000]}

rs_model = RandomizedSearchCV(estimator=model,
                              param_distributions=rs_grid,
                              n_iter=2,
                              cv=5,
                              verbose=True)

# fit the tuned model 

rs_model.fit(X_train,y_train)
```

```python
# evaluate the score 

show_scores(rs_model)
```

```python
## evalaution metric is root-mean-squraed-log-error(rmsle)

from sklearn.metrics import mean_squared_log_error,r2_score,mean_absolute_error

def rmsle(y_true,y_preds):
    """
    evalaute rmsle for true labels and predicted labels
    """
    rmsle_score = np.sqrt(mean_squared_log_error(y_true,y_preds))
    return rmsle_score

def show_scores(model):
    train_preds = model.predict(X_train)
    valid_preds = model.predict(X_valid)
    
    score = {"Training-r2-score":r2_score(y_train,train_preds),
             "valid-r2-score":r2_score(y_valid,valid_preds),
             "Training-mse":mean_absolute_error(y_train,train_preds),
             "Valid-mse":mean_absolute_error(y_valid,valid_preds),
             "Training-rmsle":rmsle(y_train,train_preds),
             "Valid-rmsle":rmsle(y_valid,valid_preds)}
    
    return score
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
bull_dozer = pd.read_csv("data/TrainAndValid.csv",
                         low_memory = False,
                         parse_dates = ["saledate"])
```

```python
bull_dozer = bull_dozer[:200000]
```

```python
bull_dozer.head()
```

```python
## arrange dates 

bull_dozer.sort_values(by=["saledate"],inplace=True,ascending=True)
```

```python
bull_dozer.saledate
```

```python
# feature engineering 

bull_dozer["saleyear"] = bull_dozer.saledate.dt.year 
bull_dozer["salemonth"] = bull_dozer.saledate.dt.month
bull_dozer["saleday"] = bull_dozer.saledate.dt.day
bull_dozer["saledayofyear"] = bull_dozer.saledate.dt.dayofyear
bull_dozer["saleweekofyear"] = bull_dozer.saledate.dt.weekofyear

## drop saledate colunmn 

bull_dozer.drop("saledate",axis=1,inplace=True)
```

```python
# check for missing values 

bull_dozer.isna().sum()
```

```python
## check for string datatype and convert them to category 

for label,content in bull_dozer.items():
    if(pd.api.types.is_string_dtype(content)):
        print(label)
        ## convert to category 
        bull_dozer[label] = content.astype('category').cat.as_ordered()
```

```python
bull_dozer.dtypes
```

```python
## fill missing values and conevert category to numeric 

for label,content in bull_dozer.items():
    if pd.api.types.is_numeric_dtype(content):
        print(label)
        # fill missing values with median 
        bull_dozer[label] = content.fillna(content.median())
    # fill categorical columns 
    if not pd.api.types.is_numeric_dtype(content):
        print(label)
        bull_dozer[label] = pd.Categorical(content).codes+1
```

```python
bull_dozer.isna().sum()
```

```python
bull_dozer.dtypes
```

```python
## choose a model 

from sklearn.ensemble import RandomForestRegressor 
model = RandomForestRegressor()

%time 

model.fit(bull_dozer.drop("SalePrice",axis=1),bull_dozer["SalePrice"])

model.score(bull_dozer.drop("SalePrice",axis=1),bull_dozer["SalePrice"])
```

```python
## get train and valid dataset 

train_set = bull_dozer[bull_dozer.saleyear==2007]
valid_set = bull_dozer[bull_dozer.saleyear!=2007]
len(train_set),len(valid_set)
```

```python
## split into X train and y train 

X_train,y_train = train_set.drop("SalePrice",axis=1),train_set["SalePrice"]
X_valid,y_valid = valid_set.drop("SalePrice",axis=1),valid_set["SalePrice"]

len(X_train),len(y_train),len(X_valid),len(y_valid)
```

```python
## fit the model 

%time 
model.fit(X_train,y_train)
```

```python
## evaluation matrix rmsle 

from sklearn.metrics import mean_squared_log_error,r2_score,mean_absolute_error,mean_squared_error

def rmsle(y_true,y_preds):
    return np.sqrt(mean_squared_log_error(y_true,y_preds))

def show_scores(model):
    train_preds = model.predict(X_train)
    valid_preds = model.predict(X_valid)
    
    scores = {"r2_train":r2_score(y_train,train_preds),
              "r2_valid":r2_score(y_valid,valid_preds),
              "mse_train":mean_absolute_error(y_train,train_preds),
              "mse_valid":mean_absolute_error(y_valid,valid_preds),
              "msse_train":mean_squared_error(y_train,train_preds),
              "msse_valid":mean_squared_error(y_valid,valid_preds),
              "rmsle_train":rmsle(y_train,train_preds),
              "rmsle_valid":rmsle(y_valid,valid_preds)}
    return scores  

show_scores(model)
```
