### NOW WHICH PLOTTING METHOD SHOULD WE USE NOW?(pyplot method the above one or Object oriented method that we done before)

1. when plotting something quickly use pyplot 
2. when plotting something advance use OOP method

```python
over_50 = heart_disease[heart_disease["age"]>50]
```

```python
len(over_50)  ##total patients above age 50 having heart disease
```

```python
over_50.head(10)
```

```python
#Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');   #c stands for cover the target column
```

```python
# OO method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
```

```python
## OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
# customize the plot
 
ax.set(title='Heart disease and cholesterol levels',
       xlabel='age',
       ylabel='cholesterol');

# add a legend ie how we write index in a graph here is called legend

ax.legend(*scatter.legend_elements(),title = 'target');

# Add a horizontal line

#ax.axhline(over_50["chol"].mean());  

# add a line style

ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

In this way we can customize our own plots

```python
over_50.head(10)
```

```python
# subplots of age thalach and cholestrol will have same x axis as age and different y axis named as cholesterol and thalach

fig,((ax0,ax1)) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10))
#add data
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
#customize ax0
ax0.set(title="Heart disease with cholesterol levels",
        xlabel = 'age',
        ylabel = 'chol');

#add legend
ax0.legend(*scatter.legend_elements(),title="target");

#add a mean line
ax0.axhline(over_50["chol"].mean(),
           linestyle='--');

#now add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

#customize the plot
ax1.set(title="Heart disease and heart rates",
        xlabel="age",
        ylabel="heart-rates")

#add a legend
ax1.legend(*scatter.legend_elements(),title="target");

#add a horizontal line
ax1.axhline(y = over_50['thalach'].mean(),
            linestyle='--');
```

In this we we can plot differnt subplots using OO method
